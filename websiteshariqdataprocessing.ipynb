```python
# Code will create a graph plotting Subject1432's FECO2 and VCO2 over VO2.
# Gas exchange threshold and respiratory compensation are plotted.

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt


#read the file. header is first row. skip the 3 rows after that because they don't contain useful info
df5 = pd.read_csv('../input/vo2data/subject_1432.csv', header=[0], skiprows=[1,2,3])

# simplify our terms 
x = df5['VO2']
y = df5['FECO2']
y1 = df5['VCO2']

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(8, 10))
fig.subplots_adjust(hspace=0)

ax[0].plot(x, y, "o", label=('FECO2'), c='r', clip_on=False, zorder=1)
ax[0].legend()
ax[0].spines[['right', 'top']].set_visible(False)
ax[0].set(ylabel=('%'))

ax[1].plot(x, y1, 'o', label=('VCO2'), c='b', clip_on=False, zorder=1)
ax[1].legend()
ax[1].spines[['right', 'top']].set_visible(False)
ax[1].set(ylabel=('L/min'))
ax[1].set(xlabel=('VO2 (L/min)'))


# plot 2 vertical lines using axvline command.
# in order to get the lines crossing both subplots, we must disable clipping.
# zorder simply puts the vertical lines behind the data points if they intersect.

ax[0].axvline(x=1.98,ymin=0,ymax=1,c="green",linewidth=2,zorder=0, clip_on=False)
ax[1].axvline(x=1.98,ymin=0,ymax=1,c="green",linewidth=2, zorder=0, clip_on=False)
    
ax[0].axvline(x=3.19,ymin=0,ymax=1,c="green",linewidth=2,zorder=0, clip_on=False)
ax[1].axvline(x=3.19,ymin=0,ymax=1,c="green",linewidth=2, zorder=0, clip_on=False)

fig.savefig("VO2_VCO2_FECO2.png", dpi=300, bbox_inches = "tight")
fig.show()
```
